```python
import os
import pandas as pd
import math
import numpy as np
import random
import matplotlib.pyplot as plt
from statistics import median,mean

#input file 
input_dir = r'/Users/arun/Downloads/CS677 - PY/'
input_file  = os.path.join(input_dir, 'zip.csv')

df = pd.read_csv(input_file)
df.head(5)
```

```python
df.drop(df.columns.difference(['zipcode','agi_stub','N1']),1,inplace=True)
df2 = df.copy()
df2 = df2[((df[['zipcode']] != 0) & (df[['zipcode']] != 99999)).all(axis=1)]
df2 = df2.reset_index(drop=True)
df2.head(5)
```

```python
sumN1 = dict(df2.groupby('zipcode')['N1'].sum())
```

```python
df2 = df2.assign(Value = lambda x: x.N1 * x.agi_stub.replace({1:12500,2:37500,3:62500,4:87500,5:112500,6:212500}))
```

```python
mulN1 = dict(df2.groupby('zipcode')['Value'].sum())
```

```python
df2["agi_stub"] = df2["agi_stub"].replace({1:12500,2:37500,3:62500,4:87500,5:112500,6:212500})
```

```python
agi = {}
for x in df2['zipcode'].unique().tolist():
    agi[x] = sorted(list(df2.loc[df['zipcode'] == x, 'agi_stub']))
```

```python
final = {}
for z,s in sumN1.items():
    temp = []
    temp.append(s)
    temp.append(mulN1[z])
    for val in agi[z]:
        temp.append(val)
    final[z] = temp
    
```

```python
f = pd.DataFrame.from_dict(final, orient='index',columns = ['sum', 'total','1','2','3','4','5','6'])
f.fillna(0, inplace = True) 
f['result'] = f['total']/f['sum']
f
```
